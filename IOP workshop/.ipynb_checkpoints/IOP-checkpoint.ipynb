## Functions
Defining functions is really easy in Julia. 

Once you have defined a function, you can use this within your script or REPL environment. 

```julia
# Defining function in python: 
#   def add(a,b):
#     return a+b
#
# Defining functions in Julia:

function add(a,b)
  return a+b
end                
```

```julia
# You can use any types that the arithmetic works on
# or any variables stored as those types. 

a = 3 # float variable
b = 4. # int variable

ints = add(3, 4)
floats = add(3., 4.)
variables = add(a, b)

println(ints)   # using println() instead of print() creates a new line for each output
println(floats)
println(variables)
```

Because Julia is a dynamic typing language, we didn't have to specify the types of variables needed by our function, and so we are only limited by the types which work with arithmetic. 

It is normally good practise to specify the types you want to be taken into your functions, 

```julia
# We can also specify types: 

function static_add(a::Int64, b::Int64)
  return a+b
end 
```

```julia
a = static_add(2.,4.)
print(a)
```

## Solving the Harmonic Oscillator
The harmonic oscillator is common across many areas of physics. 

It is used as a model for classical problems of springs as well as in atomic vibrations.

The force from a spring is given by: $F = -kx$

Using Newton's second law: $F = ma$ , we set up a second order differential equation: 

> $m \ddot{x} = -k x $

rearranging gives: 
> $\ddot{x} = -\omega x$ 

where $\omega = \frac{k}{m}$


```julia
import Pkg
Pkg.add("DifferentialEquations")
```

```julia
# Set up the packages we need 
using Plots, DifferentialEquations

# Set out system parameters
m = 1.
k = 1.
ω = k/m

# Define a function for our ODE to be solved
function HarmonicOscillator(du,u,p,x)
    # Classical equation:
    # a(t) = (1/m) k^2 x
    - ω^2 * u[1]
end

# Set parameters for the initial displacement and velocity
v0 = 0.0
u0 = 1.0

# Define the time span for the problem to be solved over
tspan = (0.0,10.0)

# use the Differential Equations function from the package to define the problem and solution
prob = SecondOrderODEProblem(HarmonicOscillator,v0,u0,tspan)
sol = solve(prob)

# Plot the solution
plot(sol, xlabel = "time", ylabel = "displacement", title = "Simple Harmonic Oscilator")
```
